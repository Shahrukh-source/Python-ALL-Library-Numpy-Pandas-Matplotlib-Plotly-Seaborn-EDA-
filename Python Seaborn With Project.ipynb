```python
tips =sns.load_dataset("tips")
tips
```

# relplot

```python
sns.relplot(x ="tip", y="total_bill", data = tips)
```

```python
sns.relplot(x = 'tip' , y = 'total_bill' , data = tips , hue = 'day' , style = 'smoker',size = 'size') # hue mens color,and blue color show day but how many smoker on day so therefore smoker countity we have use <<<<style>>, size show in group ,small grp and big gro
```
